# Exp2 - For a given set of training data examples stored in a .CSV file, implement and demonstrate the Candidate-Elimination algorithm to output a description of the set of all hypotheses consistent with the training examples.

```python
import numpy as np
import pandas as pd
```

```python
# Load data [cite: 24, 25, 27]
data = pd.read_csv('enjoysport.csv')
concepts = np.array(data.iloc[:, 0:-1])
target = np.array(data.iloc[:, -1])
data
```

```python
def learn(concepts, target):
    # 2. Initialization
    specific_h = concepts[0].copy()
    # Initial General hypothesis is all '?'
    general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]

    print("Initial Specific_h:", specific_h)
    print("Initial General_h:", general_h)
    print("-" * 30)

    # 3. Iterate through each training example
    for i, h in enumerate(concepts):
        print(f"Step {i+1}: Training Example {h} ({target[i]})")

        # If example is positive [cite: 7, 38]
        if target[i].lower() == "yes":
            for x in range(len(specific_h)):
                # If feature is inconsistent, generalize S
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'
                    # Also update G to stay consistent
                    general_h[x][x] = '?'

        # If example is negative
        if target[i].lower() == "no":
            for x in range(len(specific_h)):
                # Specialization of G
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]
                else:
                    general_h[x][x] = '?'

        # 4. Display S and G at each step
        print("Specific_h:", specific_h)
        # Filter '?' rows for cleaner display of G
        current_g = [g for g in general_h if g != ['?', '?', '?', '?', '?', '?']]
        print("General_h:", current_g)
        print("-" * 30)

    return specific_h, current_g
```

```python
s_final, g_final = learn(concepts, target)

print("Final Specific_h:", s_final)
print("Final General_h:", g_final)
```
